## Filter Selection - Mutual Information
Filter Selection can use mutual information or correlation. 

```python
from sklearn.metrics import mutual_info_score

def calc_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    mi = mutual_info_score(None, None, contingency=c_xy)
    return mi

mi_df = pd.DataFrame(columns=['col_index', 'MI'])

#The MI will be calculated for every column, then appended to a dataframe
for col in data.columns:
    mi = calc_MI(target, data[col], 20)
    mi_df = mi_df.append({'col_index': col, 'MI' : mi}, ignore_index=True)
    #print("Mutual information=%.2f"%mi)
```

After calculating the mutual information for every column, we can see the top 10 attributes. 

```python
mi_df.sort_values(by='MI', ascending = False).head(12)
```

## Wrapper - Backward Model Selection
The wrapper method includes forward/backwards step-wise selection. 

```python
#using recursive feature elimination library
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

estimator = LinearRegression()
selector = RFE(estimator, 12, step=5)
selector = selector.fit(data, target)
print(selector.support_)
print(selector.ranking_)
```

The recursive feature elimination returned a list of booleans that represents whether an attribute is one of the selected 10 features. 

To get the names of the attributes, I itterated over the list and appended the indices where the value is true.

```python
wrapper = []

for val in range(len(selector.support_)):
    if (selector.support_[val] == True):
        wrapper.append(data.columns[val])

print(wrapper)
```

## Embedded -  Lasso Regression


```python
from sklearn import linear_model

clf = linear_model.Lasso(alpha = 0.005)
clf.fit(data, target)

print(clf.coef_)

print(clf.intercept_)
```

Lasso regression shows us which features were shrunk down to 0. The remaining ones will be used for feature selectoin. 

```python
lasso = []

for val in range(len(clf.coef_)):
    if (clf.coef_[val] != 0):
        lasso.append(data.columns[val])
        
print(lasso)
```
